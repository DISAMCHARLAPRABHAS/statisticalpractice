### Barchart

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
sales=[23,43,50,12,34,45,27]
days=[1,2,3,4,5,6,7]
plt.bar(days,sales)
plt.show()
```

```python
plt.bar(days,sales,color='green',edgecolor='black',label='sales')
plt.title('Day wise sales- Bar chart')
plt.xlabel('days')
plt.ylabel('sales')
plt.grid()
plt.legend()
```

### linechart

```python
plt.plot(days,sales,color='green', marker='o',label='sales',lw=2,markersize=4)
#lw=linewidth
plt.title('Day wise sales- Bar chart')
plt.xlabel('days')
plt.ylabel('sales')
plt.grid()
plt.legend()
```

### scatter chart

```python
s1 = np.random.randint(20,30,50)
s2 = np.random.randint(20,30,50)
```

```python
plt.scatter(s1,s2,marker='o',color='maroon',s=80)
plt.title('s1 vs s2 scatter chart')
plt.xlabel('s1')
plt.ylabel('s2')
plt.show()
```

### histogram

```python
x = np.random.choice([1,2,3,4,5,6,7],size=20)
print(x)
```

```python
from collections import Counter
xc=Counter(x)
print(xc)
```

```python
plt.hist(x)
plt.show()
```

```python
plt.hist(x,bins=7,edgecolor='black',color='red')
plt.show()
```

```python
print(xc)
plt.hist(x,bins=6,edgecolor='black',color='red')
plt.show()
```

### piechart

```python
marks=[76,23,91,50,59,69]
sub=['maths','phython','java','c++','javascript','html']
```

```python
per_DS=(marks[2]/sum(marks)*100)
print(per_DS)
```

```python
plt.pie(marks,labels=sub,autopct='%.2f%%')
plt.show()
```
